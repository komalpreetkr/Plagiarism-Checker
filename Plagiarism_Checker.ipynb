```python
# Step 1: Install required library
!pip install scikit-learn

# Step 2: Import libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

# Step 3: Upload two text files
print("📤 Upload the ORIGINAL and STUDENT text files (e.g., original.txt and student.txt):")
uploaded = files.upload()
filenames = list(uploaded.keys())

# Step 4: Read the uploaded files
with open(filenames[0], 'r', encoding='utf-8') as file1:
    text1 = file1.read()

with open(filenames[1], 'r', encoding='utf-8') as file2:
    text2 = file2.read()

# Step 5: Calculate similarity using TF-IDF + Cosine Similarity
texts = [text1, text2]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

# Step 6: Show Result
percentage = round(similarity * 100, 2)
print(f"✅ Similarity: {percentage}%")

# Optional flag
if percentage > 80:
    print("⚠️ High similarity! Possible plagiarism.")
elif percentage > 50:
    print("🔍 Medium similarity. Review recommended.")
else:
    print("✅ Low similarity. Likely original.")
```
